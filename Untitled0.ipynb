```python

sentence = "the movie was good"
tokens=sentence.split()

hidden_state = {
    "topic": None,
    "negation_active": False,
    "sentiment_word": None
}

for token in tokens:
  if token in ["movie", "food", "product"]:
    hidden_state["topic"] = token

  if token == "not":
    hidden_state["negation_active"] = True

  if token in ["good", "bad", "great", "terrible"]:
    hidden_state["sentiment_word"] = token

print("Token:", token)
print("Hidden State:",  hidden_state)
print("----------")
```

```python
import tensorflow as tf
from tensorflow.keras import layers

vocab_size = 1000
embedding_dim= 16
sequence_lenghts= 8

simple_rnn_model = tf.keras.Sequential([
    layers.Embedding(input_dim=vocab_size, output_dim= embedding_dim, input_length=sequence_lenghts),
    layers.SimpleRNN(32),
    layers.Dense(1, activation= "sigmoid")
])

simple_rnn_model.summary()
```
